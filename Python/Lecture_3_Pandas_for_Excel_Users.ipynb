# Introduction to Pandas for Excel Users

Welcome to this introduction to Pandas, a powerful data analysis library in Python. If you're an Excel user who has never used Python before, this notebook is designed just for you!

### What is Pandas?
Pandas is an open-source library in Python, widely used for data analysis and manipulation. It is somewhat similar to Microsoft Excel but has many more powerful features for working with data programmatically. You can perform operations like reading data, cleaning it, manipulating it, and visualizing it.

Throughout this notebook, we'll illustrate how you can perform common data analysis tasks with Pandas and compare them with equivalent Excel operations.

````python
# Import necessary libraries
import pandas as pd
import yfinance as yf
````

## Fetching Stock Data from Yahoo Finance

We will use the Yahoo Finance API to fetch some stock data for demonstration. This data includes information like stock prices, volume, and more. If you are familiar with importing data into Excel, this is similar to getting data from an external data source.

The Yahoo Finance API can be accessed using the `yfinance` package. Letâ€™s fetch some data for a popular stock: Apple Inc. (AAPL).

````python
# Fetch stock data for Apple (AAPL)
aapl_data = yf.download('AAPL', period='max') # start='2022-01-01', end='2023-01-01')
print(type(aapl_data))
# Display the first few rows of data
````

````python
aapl_data.head(10)
````

## Basic Operations with Pandas

In Excel, you might use features like **Filters**, **Formulas**, and **Pivot Tables** to analyze your data. Pandas has similar functionalities but with much greater flexibility.

### 1. Viewing Data
You can view the first few rows of a DataFrame using the `.head()` method.

```python
aapl_data.head()
```

You can also view the last few rows using the `.tail()` method.

```python
aapl_data.tail()
```

### 2. Checking the Data Types
In Excel, data types are usually inferred when you enter data into cells. In Pandas, you can check the types of columns using the `.dtypes` attribute.

```python
aapl_data.dtypes
```

This helps to understand the kind of data you're working with: numerical, categorical, date, etc.


````python
# Viewing last few rows
print(aapl_data.tail())

# Checking data types
aapl_data.dtypes
````

### 3. Filtering Data
In Excel, you can filter data using dropdown menus. In Pandas, you can do the same programmatically.

Let's filter the data to show only the days where the closing price was greater than $150.

```python
filtered_data = aapl_data[aapl_data['Close'] > 150]
filtered_data.head()
```


````python
(1 * (aapl_data['Close'] > 150)).loc['2005-01-01':].plot()
````

````python
# Filtering data where closing price is greater than $150
filtered_data = aapl_data[aapl_data['Close'] > 150]
filtered_data.head()
````

### 4. Descriptive Statistics
In Excel, you may use functions like `AVERAGE()`, `SUM()`, `MIN()`, `MAX()` to get summary statistics. In Pandas, you can easily get these statistics for all columns using the `.describe()` method.

```python
aapl_data.describe()
```


````python
# Get descriptive statistics
aapl_data.describe()
````

## Data Visualization
One of the strengths of using Pandas and Python is the ability to easily visualize your data. You can use the `.plot()` method to quickly generate plots.

Let's plot the closing price of the stock over time:

```python
aapl_data['Close'].plot(title='Apple Stock Closing Prices (2022)')
```


````python
# Plotting the closing prices
aapl_data['Close'].plot(title=f'Apple Stock Closing Prices as of {aapl_data.index[-1]}', figsize=(10, 6))
````

````python
# how does Pandas know how to interpret the dates?
# It uses a special data formal, pd.datetime
aapl_data.index[100]
````

````python
pd.to_datetime('2022-01-01')
````

## Saving Data to Excel
After performing data manipulations, you might want to save your results back to an Excel file. Pandas makes it easy to do so using the `.to_excel()` method.

```python
filtered_data.to_excel('filtered_aapl_data.xlsx', index=False)
```

This will save the filtered data to an Excel file named `filtered_aapl_data.xlsx`.


````python
# Save filtered data to Excel
filtered_data.to_excel('filtered_aapl_data.xlsx', index=False)
````
