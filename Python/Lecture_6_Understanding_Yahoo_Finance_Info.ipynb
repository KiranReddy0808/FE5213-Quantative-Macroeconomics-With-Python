
# Explanations of `stock_data.info` Attributes in Yahoo Finance API

In this notebook, we will use the `yfinance` library to fetch the `info` attribute of a stock and explain the meaning of each key-value pair returned by this method. We'll use **Apple Inc. (AAPL)** as an example.

## 1. Importing Libraries

Let's start by importing the necessary libraries.


```python

# Import necessary libraries
import yfinance as yf
```


## 2. Fetching `info` for Apple Inc. (AAPL)

We will fetch the `info` attribute, which provides detailed information about the stock.


```python

# Define stock symbol for Apple Inc.
stock_data = yf.Ticker("AAPL")

# Fetch and display the info
info = stock_data.info
info
```


## 3. Explanation of Common `info` Attributes

Here are explanations for some of the most common fields returned by `info`:

### 3.1 `sector`
- The business sector the company operates in.
- **Example**: "Technology"

### 3.2 `industry`
- The industry within the sector.
- **Example**: "Consumer Electronics"

### 3.3 `marketCap`
- The market capitalization of the company (total market value of its outstanding shares).
- **Example**: 2,400,000,000,000 (this is in dollars, for $2.4 trillion)

### 3.4 `fullTimeEmployees`
- The number of full-time employees working for the company.
- **Example**: 147,000

### 3.5 `longBusinessSummary`
- A summary description of the business and its operations.
- **Example**: "Apple Inc. designs, manufactures, and markets smartphones, personal computers, tablets, and more."

### 3.6 `city`, `state`, `country`
- The location of the company’s headquarters.
- **Example**: "Cupertino", "California", "United States"

### 3.7 `currency`
- The currency in which the stock is traded.
- **Example**: "USD" (United States Dollar)

### 3.8 `dividendYield`
- The dividend yield is a financial ratio that shows how much a company pays out in dividends each year relative to its stock price.
- **Example**: 0.0065 (0.65%)

### 3.9 `dividendRate`
- The dollar amount paid in dividends per share annually.
- **Example**: 0.88 (This means $0.88 per share annually)

### 3.10 `trailingPE`
- The price-to-earnings (P/E) ratio based on the last 12 months of earnings.
- **Example**: 30.12

### 3.11 `forwardPE`
- The P/E ratio based on future earnings projections.
- **Example**: 27.24

### 3.12 `profitMargins`
- The profit margin of the company, indicating how much of each dollar of sales is profit.
- **Example**: 0.23 (This means 23% profit margin)

### 3.13 `priceToSalesTrailing12Months`
- The price-to-sales (P/S) ratio, a valuation ratio that compares the company's stock price to its revenues.
- **Example**: 7.10

### 3.14 `beta`
- A measure of the stock's volatility relative to the market. A beta greater than 1 indicates higher volatility than the market.
- **Example**: 1.20

### 3.15 `fiftyTwoWeekHigh`
- The highest price at which the stock has traded in the last 52 weeks.
- **Example**: 150.00 (This means the stock’s 52-week high price was $150)

### 3.16 `fiftyTwoWeekLow`
- The lowest price at which the stock has traded in the last 52 weeks.
- **Example**: 100.00 (This means the stock’s 52-week low price was $100)

### 3.17 `twoHundredDayAverage`
- The average stock price over the last 200 days.
- **Example**: 135.24

### 3.18 `fiftyDayAverage`
- The average stock price over the last 50 days.
- **Example**: 140.12

### 3.19 `volume`
- The number of shares traded on the most recent trading day.
- **Example**: 20,000,000

### 3.20 `averageVolume`
- The average number of shares traded per day over a given period, typically 3 months.
- **Example**: 30,000,000

### 3.21 `pegRatio`
- The price-to-earnings growth ratio, which considers earnings growth. A lower PEG ratio is considered more attractive.
- **Example**: 1.90

### 3.22 `epsTrailingTwelveMonths`
- The company's earnings per share over the past 12 months.
- **Example**: 4.50

### 3.23 `epsForward`
- The projected earnings per share for the next fiscal year.
- **Example**: 5.10

### 3.24 `sharesOutstanding`
- The total number of shares that are currently owned by all shareholders.
- **Example**: 16,000,000,000 (16 billion shares outstanding)

### 3.25 `floatShares`
- The number of shares available for trading by the public.
- **Example**: 15,500,000,000

### 3.26 `heldPercentInstitutions`
- The percentage of shares held by institutional investors.
- **Example**: 0.59 (59% of shares held by institutions)

### 3.27 `shortRatio`
- The ratio of shares that are being shorted to the average daily trading volume.
- **Example**: 1.2 (This means the short interest is 1.2 times the daily average volume)

### 3.28 `forwardEps`
- The expected earnings per share for the next year.
- **Example**: 5.50

### 3.29 `pegRatio`
- The price/earnings-to-growth ratio, used to determine stock valuation while factoring in growth. A lower PEG ratio is better.
- **Example**: 2.0

### 3.30 `sharesShort`
- The total number of shares currently being shorted.
- **Example**: 45,000,000 (45 million shares shorted)

## Conclusion

These are the most commonly used attributes of the `info` dictionary returned by `yfinance`. You can explore additional fields by examining the full output of `stock_data.info`.
