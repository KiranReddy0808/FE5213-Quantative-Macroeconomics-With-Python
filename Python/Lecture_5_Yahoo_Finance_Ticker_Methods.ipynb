
# Exploring Yahoo Finance Ticker Methods with yfinance

This notebook demonstrates how to use the `yfinance` library to fetch stock data and provides detailed explanations and examples of all major methods available for the `yf.Ticker()` object.

We will use **Apple Inc. (AAPL)** as an example and cover the following attributes and methods:

1. General Information
2. Historical Data
3. Dividends & Splits
4. Analyst Recommendations
5. Earnings & Financial Data
6. Shareholders Information
7. Options Data

## 1. Importing Libraries

Let's start by importing the necessary libraries.


```python

# Import necessary libraries
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
```


## 2. Fetching Stock Data for Apple Inc.

We will fetch data for **Apple Inc. (AAPL)** and explore its attributes.


```python

# Define stock symbol for Apple Inc.
stock_data = yf.Ticker("AAPL")
```


## 3. General Information: `info`

The `info` attribute provides general information about the company, including its sector, market cap, full-time employees, and much more.


```python

# Fetch general information about the stock
info = stock_data.info
info
```


## 4. Historical Market Data: `history()`

The `history()` method returns historical market data, including Open, High, Low, Close prices, Volume, Dividends, and Stock Splits.


```python

# Fetch historical market data for Apple Inc.
history_data = stock_data.history(period="5y")

# Display the first few rows
history_data.head()
```

```python
history_data[history_data['Dividends'] > 0].head()
```


### Visualizing Historical Data

We can plot the historical **Close** prices to visualize Apple's stock performance.


```python

# Plot the historical Close prices
plt.figure(figsize=(10,6))
plt.plot(history_data['Close'], label="Close Price")
plt.title('AAPL Stock Price (5-Year)')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.grid(True)
plt.legend()
plt.show()
```


## 5. Dividends: `dividends`

The `dividends` attribute provides the dividend history of the stock.


```python

# Fetch dividend history for Apple Inc.
dividends = stock_data.dividends

# Display dividend history
dividends.head()
```


### Visualizing Dividend Payments

We can also visualize the dividend history.


```python

# Plot the dividend history
plt.figure(figsize=(10,6))
plt.bar(dividends.index, dividends.values, label="Dividends", color="orange")
plt.title('AAPL Dividend History')
plt.xlabel('Date')
plt.ylabel('Dividend Amount (USD)')
plt.grid(True)
plt.legend()
plt.show()
```


## 6. Stock Splits: `splits`

The `splits` attribute shows the stock split history. Stock splits increase the number of shares while lowering the share price proportionally.


```python

# Fetch stock splits for Apple Inc.
splits = stock_data.splits
splits
```


## 7. Corporate Actions: `actions`

The `actions` attribute combines both **dividends** and **splits** into one DataFrame.


```python

# Fetch corporate actions (dividends and splits)
actions = stock_data.actions
actions
```


## 8. Analyst Recommendations: `recommendations`

The `recommendations` attribute provides a summary of analyst recommendations for the stock, such as "Buy", "Hold", or "Sell".


```python

# Fetch analyst recommendations for Apple Inc.
recommendations = stock_data.recommendations
recommendations.tail()
```


## 9. Earnings Data: `earnings`

The `earnings` attribute provides quarterly and annual earnings reports.


```python

# Fetch annual earnings data for Apple Inc.
earnings = stock_data.income_stmt
earnings
```


## 10. Cash Flow: `cashflow`

The `cashflow` attribute provides detailed information on the cash flow of the company.


```python

# Fetch cash flow data for Apple Inc.
cashflow = stock_data.cashflow
cashflow
```


## 11. Balance Sheet: `balance_sheet`

The `balance_sheet` attribute provides information about the company's assets, liabilities, and shareholders' equity.


```python

# Fetch balance sheet data for Apple Inc.
balance_sheet = stock_data.balance_sheet
balance_sheet
```


## 12. Financials: `financials`

The `financials` attribute provides the income statement of the company.


```python

# Fetch financials (income statement) for Apple Inc.
financials = stock_data.financials
financials
```


## 13. Major Shareholders: `major_holders`

The `major_holders` attribute shows the major shareholders of the company.


```python

# Fetch major shareholders for Apple Inc.
major_holders = stock_data.major_holders
major_holders
```


## 14. Institutional Holders: `institutional_holders`

The `institutional_holders` attribute shows the institutions holding shares in the company.


```python

# Fetch institutional holders for Apple Inc.
institutional_holders = stock_data.institutional_holders
institutional_holders
```


## 15. Options: `options` and `option_chain()`

The `options` attribute lists the available expiration dates for options. The `option_chain()` method fetches the option chain for a specific date.


```python

# Fetch available option expiration dates for Apple Inc.
options_dates = stock_data.options
options_dates
```

```python

# Fetch option chain for the first expiration date
option_chain = stock_data.option_chain(options_dates[0])
option_chain.calls.head()
```
